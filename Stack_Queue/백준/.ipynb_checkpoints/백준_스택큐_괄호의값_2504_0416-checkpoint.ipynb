```python
# () : 2, [] : 3
# 값(XY)= 값(X)+값(Y) 
# ‘(X)’ 의 괄호값은 2×값(X)
# ‘[X]’ 의 괄호값은 3×값(X) 
# (()[[]])([]) -> 28

'''
해결 : 스택


괄호의 길이 1~ 30
1. 올바른 괄호 문자열 인가, 아니라면 0 출력
2. 계산하기

'(' 이 들어오면,
stack에 push하고
변수(temp)에 2를 곱한다.
'[' 가 들어오면,
stack에 push하고
변수(temp)에 3을 곱한다.
')' 이 들어오면,
if stack이 empty하거나 stack의 top이 ')' 가 아니면, wrong input.
else 만약 바로 전의 문자가 '('면 result 변수에 더해주고, temp를 2로 나눠준다.
바로 전의 문자가 '(' 아니면 그냥 temp를 2로 나눈다.
'''
```

```python
s = input()

# s = list(s)
stack1 = []
stack2 = []
temp = 1
result = 0
check = True

for i in range(len(s)):
    if s[i] =='(':
        stack1.append(s[i])
  
        temp *= 2
    elif s[i] =='[':
        stack2.append(s[i])
        temp *= 3
       
        
    elif s[i] == ')':
        if stack1:
            if s[i-1] == '(':
                result += temp
            stack1.pop()
            temp//=2
        else:
            check=False
            break
        
    elif s[i] == ']':
        if stack2:
            if s[i-1] == '[':
                result += temp
            stack2.pop()
            temp//=3
        else:
            check=False
            break
            
if not stack1 and not stack2 and check==True:
    print(result)
else:
    print(0)
```
