```python
# bfs

from collections import deque
n,m = map(int, input().split())

graph = [list(map(str, input())) for _ in range(m)]
visited = [[0]*n for _ in range(m)]
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs():
    q = deque()
    q.append((i, j))
    visited[i][j] =1  # 삽입 후 방문 처리 꼭 하기
    
    cnt = 1
    while q:
        x, y = q.popleft()
        
        for t in range(4):
            nx = x + dx[t]
            ny = y + dy[t]
            if nx >= 0 and  nx < m and ny >=0 and ny<n and graph[nx][ny]== graph[x][y] and visited[nx][ny]==0:
                q.append((nx, ny))
                visited[nx][ny] =1
                cnt += 1
    return cnt

W, B = 0, 0
for i in range(m):
    for j in range(n):
        if visited[i][j] == 0:
            result = bfs()
            if graph[i][j] == "W":
                W += result ** 2
            else:
                B += result ** 2
print(W, B)
```

```python
# 뱀 2 
# 사과 3
# 벽 1
# 빈 공간 0
'''
6
3
3 4
2 5
5 3
3
3 D
15 L
17 D   ->   9
'''

n = int(input()) # 보드 개수
k = int(input()) # 사과 개수
graph = [[1]*(n+2)] + [[1]+ [0]*n+ [1] for _ in range(n)] + [[1]*(n+2)] # 그래프 만드는 법

for _ in range(k): # 사과 위치
    a, b = map(int, input().split())
    graph[a][b] = 3 # 사과 위치 -> 3

times = [] # 시간과 방향 정보
l = int(input())
for _ in range(l):
    x, c = map(str, input().split())
    times.append((int(x), c))
    
# 시계 방향 : 동 0 남 1 서 2 북 3 (오른쪽에서 시작함)
dx = [0, 1, 0, -1]
dy = [1, 0, -1, 0]

def turn(direction, c): # direction은 dx, dy의 index
    if c == "L":
        direction = (direction - 1) % 4
    else:
        direction = (direction + 1) % 4
    return direction

def go():
    x, y = 1, 1 # 뱀 머리
    graph[x][y] = 2 # 뱀 위치 -> 2
    direction = 0  # direction은 dx, dy의 index
    time = 0 # 시작한 뒤 초 시간
    next_turn = 0 # x초 뒤에 회전 정보
    q = [(x, y)] # 뱀이 차지한 위치 (꼬리가 앞일 것임)
    
    while True:
        
        nx = x + dx[direction]
        ny = y + dy[direction]
        
        if 1 <= nx and nx <= n and 1<= ny and ny <= n and graph[nx][ny] != 2 :
            # 사과 없다면 이동 후 꼬리 제거하기
            if graph[nx][ny] == 0:
                graph[nx][ny] = 2
                q.append((nx, ny))
                tx, ty = q.pop(0) #꼬리 제거
                graph[tx][ty] = 0
            # 사과 있다면 사과 제거, 꼬리 그대로
            if graph[nx][ny] == 3:
                graph[nx][ny] = 2
                q.append((nx, ny))
        else: # 벽이랑 부딪혔다면
            time += 1
            break
        x, y = nx, ny
        time += 1
        if next_turn < l and time == times[next_turn][0]: # 회전할 시간이 되면 회전하기
            direction = turn(direction, times[next_turn][1])
            next_turn += 1
    return time
print(go())
            
                



```
