```python
from collections import deque, defaultdict
import math
import sys
sys.setrecursionlimit(10**6)

def bfs(x, y, k, visited, height, n, land):
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1 ,1]
    
    queue = deque()
    queue.append((x,y))
    visited[x][y] =k
    
    while queue: 
        x, y = queue.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx < n and 0 <= ny < n and visited[nx][ny] == 0:
                if abs(land[x][y] - land[nx][ny]) <= height:
                    queue.append((nx, ny))
                    visited[nx][ny] = k
      
        
def get_distance(visited, height, maps, table, n):
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    for x in range(n):
        for y in range(n):
            nx = x + 1
            ny = y + 1
            # 상하 비교
            if nx < n and visited[x][y] != visited[nx][y]:
                a = min(visited[x][y], visited[nx][y])
                b = max(visited[x][y], visited[nx][y])
                table[(a,b)] = min(table[(a,b)], abs(maps[nx][y]- maps[x][y]))
            # 좌우 비교
            if ny < n and visited[x][y] != visited[x][ny]:
                a = min(visited[x][y], visited[x][ny])
                b = max(visited[x][y], visited[y][ny])
                table[(a,b)] = min(table[(a,b)], abs(maps[x][y]- maps[x][ny]))
                
def find(parent, x):
    if parent[x] != x:
        parent[x] = find(parent, parent[x])
    return parent[x]
                
def union(parent, x, y):
    x = find(parent, x)
    y = find(parent, y)
    parent[y] = x

        
def solution(land, height):
    answer = 0
    n = len(land)
    visited = [[0]*n for _ in range(n)]
    
    k =1
    for i in range(n):
        for j in range(n):
            if visited[i][j] == 0:
                bfs(i, j, k, visited, height, n, land)
                k +=1
    # land별 최솟값 구하기
    # table : {(x,y):value} x그룹->y그룹 갈때 필요한 사다리 비용 
    # defaultdict : 없는 key에 대해 호출 가능
    table = defaultdict(lambda: math.inf)#최소거리 저장 위해 dic생성하고, 기본값 무한대
    get_distance(visited, height, land, table, n)
    table = sorted(table.items(), key = lambda x: x[1])
    
    nodes = {i:i for i in range(1, k)}
    for (a,b), value in table:
        # 사다리 연결
        if find(nodes, a) != find(nodes, b):
            union(nodes, a, b)
            answer += value
        
        # 모든 지형 연결되어있는가
        if len(nodes.values()) == 1:
            return answer
    return answer

solution([[1, 4, 8, 10], [5, 5, 5, 5], [10, 10, 10, 10], [10, 10, 10, 20]], 3)
```
